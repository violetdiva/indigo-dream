```python
def function(file):
    lines = []
    for line in file:
        lines.append(line)
    return lines
with open('Data_Science_Instructor_Position_Description_US.txt', encoding = "latin-1", mode = 'r') as f: #open the file
    dsi = function(f) #put the lines to a variable (list).
    print(dsi)
```

```python
def compare_files(file1, file2):
    for line1, line2 in zip(file1, file2):
       if line1 != line2:
           print(f"Difference found:\nFile1: {line1}File2: {line2}")
```

```python
from difflib import Differ
from numpy import diff
import filecmp
```

```python
def compare_files_1(file_path1, file_path2):
    """Compares two text files line by line and prints differences."""
    try:
        for line1, line2 in zip(file_path1, file_path2):
            if line1 != line2:
               print(f"Difference found:")
               print(f"File 1: {line1.rstrip()}")
               print(f"File 2: {line2.rstrip()}")
    except FileNotFoundError:
        print("Error: One or both files not found.")
```

```python
compare_files_1(dsi, adsi)
```
