#Word Frequency

Let's start with a total number of words

```python
count = 0

# Opens a file in read mode
f = open("/content/HappyFileText copy.txt", "r")

# Gets each line till end
for line in f:
    # Splits into words
    word = line.split(" ")
    # Counts each words
    count += len(word)

print("Total Number of Words: " + str(count))
f.close()
```

Cool. Now to count individual words

```python
# Open the file in read mode
text = open("HappyFileText copy.txt", "r")

# Create an empty dictionary
d = dict()

# Loop through each line of the file
for line in text:
    # Remove the leading spaces and newline character
    line = line.strip()

    # Convert the characters in line to
    # lowercase to avoid case mismatch
    line = line.lower()

    # Split the line into words
    words = line.split(" ")


    # Iterate over each word in line
    for word in words:
        # Check if the word is already in dictionary
        if word in d:
            # Increment count of word by 1
            d[word] = d[word] + 1
        else:
            # Add the word to dictionary with count 1
            d[word] = 1

# Print the contents of dictionary
for key in list(d.keys()):
    print(key, ":", d[key])
```

# Let's try a bar chart

```python
from collections import Counter

def plot_word_frequency(file_path, top_n=10):
    """
    Plots the frequency of the top N words from a text file.

    Args:
        file_path (str): Path to the text file.
        top_n (int): Number of top words to display.
    """
    try:
        with open(file_path, 'r') as file:
            text = file.read()
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return

    words = text.lower().split()
    word_counts = Counter(words)

    most_common_words = word_counts.most_common(top_n)

    words, counts = zip(*most_common_words)

    plt.figure(figsize=(10, 6))
    plt.bar(words, counts)
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.title(f"Top {top_n} Word Frequencies")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()


file_path = "/content/HappyFileText copy.txt"
plot_word_frequency(file_path)
```

```python
from nltk.corpus import stopwords
import string
from collections import Counter
import matplotlib.pyplot as plt

def plot_word_frequency(file_path, num_words=10):
    """
    Plots the frequency of the most common words in a text file, excluding stopwords.

    Args:
        file_path (str): The path to the text file.
        num_words (int): The number of top words to display in the bar chart.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return

    text = text.lower()
    text = ''.join([char for char in text if char not in string.punctuation])
    words = text.split()

    stop_words = set(stopwords.words('english'))
    custom_stopwords = ['skillsoft', 'joy', 'hopkins','instructor', 'Joy', 'Joy Hopkins', 'Hopkins', 'Thank', 'class', 'time', 'really', 'â', 'topic','topics', 'course', 'session', 'bootcamp', 'Eian', 'question', 'questions']
    stop_words.update(custom_stopwords)

    words = [word for word in words if word not in stop_words]

    word_counts = Counter(words)
    most_common_words = word_counts.most_common(num_words)

    word_labels, word_frequencies = zip(*most_common_words)

    plt.figure(figsize=(10, 6))
    plt.bar(word_labels, word_frequencies)
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.title(f"Top {num_words} Word Frequencies (Excluding Stopwords)")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

```python
# Example usage:
file_path = "/content/HappyFileText copy.txt"
plot_word_frequency(file_path, 20)
```

```python
def create_word_treemap(file_path):
    """
    Generates and displays a treemap of word frequencies from a text file.

    Args:
        file_path (str): The path to the text file.
    """
    word_counts = {}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                word_counts[word] = word_counts.get(word, 0) + 1

    word_list = list(word_counts.keys())
    count_list = list(word_counts.values())

    plt.figure(figsize=(12, 8))
    squarify.plot(sizes=count_list, label=word_list, alpha=.8)
    plt.axis('off')
    plt.title('Word Frequency Treemap')
    plt.show()
```

```python
def create_treemap(file_path, limit = 8, stop_words_language='english'):
    """
    Generates a treemap of word frequencies from a text file, excluding stopwords.

    Args:
        file_path (str): The path to the text file.
        stop_words_language (str, optional): The language of stopwords to exclude. Defaults to 'english'.
    """
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return
    except Exception as e:
         print(f"An error occurred while reading the file: {e}")
         return

    words = word_tokenize(text.lower())
    stop_words = stopwords1

    filtered_words = [word for word in words if word.isalnum() and word not in stop_words]

    word_counts = {}
    for word in filtered_words:
        word_counts[word] = word_counts.get(word, 0) + 1

    sorted_word_counts = dict(sorted(word_counts.items(), key=lambda item: item[1], reverse=True))

    labels = list(sorted_word_counts.keys())
    sizes = list(sorted_word_counts.values())

    plt.figure(figsize=(16, 12))
    squarify.plot(sizes=sizes, label=labels, alpha=.8)
    plt.axis('off')
    plt.title('Treemap of Words (Excluding Stopwords)')
    plt.show()


file_path = 'HappyFileText copy.txt'
create_treemap(file_path, limit = 8)
```
