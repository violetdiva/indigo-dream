```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('/content/Iris.csv')
```

```python
iris.columns
```

```python
penguin = pd.read_csv('/content/penguins.csv')
```

```python
penguin.columns
```

Slide 73, pie chart example

```python
counts = penguin['species'].value_counts()
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Penguin Species')
plt.show()
```

Slide 75, scatterplot example

```python
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title("Sepal Width (cm) by Sepal Length (cm)", fontsize = "22")
plt.xlabel("Sepal Length (cm)", fontsize = "18")
plt.ylabel("Sepal Width (cm)", fontsize = "18")
plt.show()
```

Slide 76, scatterplot example (how to read a scatterplot)

```python
sns.scatterplot(data=penguin, x='bill_length_mm', y='bill_depth_mm')
plt.title("Bill Depth (mm) by Bill Length (mm)", fontsize = "22")
plt.xlabel("Bill Length (mm)", fontsize = "18")
plt.ylabel("Bill Depth (mm)", fontsize = "18")
plt.show()
```

Slide 77, Bubble chart example

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=penguin, x='bill_length_mm', y='bill_depth_mm', hue='species', size= 'flipper_length_mm', sizes = (10, 300))
plt.title("Bill Depth (mm) by Bill Length (mm)", fontsize = "22")
plt.xlabel("Bill Length (mm)", fontsize = "18")
plt.ylabel("Bill Depth (mm)", fontsize = "18")
plt.show()
```

Slide 83, Histogram example

```python
sns.histplot(data=iris, x="PetalLengthCm")
plt.title("Frequency of Petal Length (cm)", fontsize = "22")
plt.xlabel("Petal Length (cm)", fontsize = "18")
plt.ylabel("Frequency", fontsize = "18")
plt.show()
```
