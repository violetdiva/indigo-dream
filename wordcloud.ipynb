```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

Load text. Am I using the original spreadsheets or the word doc?

Let's start with the word doc, converted to a text file.

```python
def function(file):
    lines = []
    for line in file:
        lines.append(line)
    return lines
with open('/content/HappyFileText copy.txt', encoding = "latin-1", mode = 'r') as f: #open the file
    happy = function(f) #put the lines to a variable (list).
    print(happy)
```

```python
string = " ".join(happy)
```

```python
wc = WordCloud().generate(string)
plt.imshow(wc)
```

Cool. Next up, start playing with stopwords & then aesthetics.

WordCloud( ) contains parameters including the following:

  ** background_color: Color of background
  ** max_words: The maximum number of unique words used
  ** stopwords: A stopword list to exclude the words you don’t wish to display
  ** colormap: The color theme
  ** width: The width of the WordCloud image
  ** height: The height of the WordCloud image





```python
stopwords = set(STOPWORDS)
stopwords.update(['instructor', 'Joy', 'Joy Hopkins', 'Hopkins', 'Thank', 'class', 'time', 'really', 'â', 'topic', 'course', 'session', 'bootcamp', 'Eian'])
```

```python
wordcloud1 = WordCloud(background_color='white', colormap = 'twilight',
      stopwords=stopwords, width = 1200, height = 800).generate(string)
plt.imshow(wordcloud1)
```
